```python
# format columns
df.columns = df.columns.str.lower().str.replace(' ', '_')
#rename columns purchase_amount to amount
df.rename(columns={'purchase_amount_(usd)': 'purchase_amount_usd'}, inplace=True)
```

```python
df.columns
```

```python
# new columns age_group
labels = ['Young', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.cut(df['age'], bins=[0, 18, 35, 55, 100], labels=labels)
```

```python
df[['age', 'age_group']].head(100)
```

```python
# kolom rata-rata pembelian per hari
frequency_mapping = {
    'Fortnightly': 14,
    'Monthly': 30,
    'Weekly': 7,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_per_day'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['frequency_of_purchases', 'purchase_frequency_per_day']].head(100)
```

```python
df[['discount_applied', 'promo_code_used']].head(100)
```

```python
# hapus kolom promo_code_used karna pada dasarnya sama dengan discount_applied
df = df.drop(columns=['promo_code_used'])
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pilih hanya kolom numerik
numerical_cols = df.select_dtypes(include=['number']).columns

# Menentukan ukuran figure
fig, axes = plt.subplots(nrows=len(numerical_cols), ncols=2, figsize=(12, 4 * len(numerical_cols)))

for i, col in enumerate(numerical_cols):
    # Histogram dengan KDE
    sns.histplot(data=df, x=col, kde=True, ax=axes[i, 0])
    axes[i, 0].set_title(f'Histogram {col}')

    # Boxplot
    sns.boxplot(x=df[col], ax=axes[i, 1])
    axes[i, 1].set_title(f'Boxplot {col}')

# Sesuaikan tata letak agar lebih rapi
plt.tight_layout()
plt.show()
```

```python
# Tabel jumlah predikat per mata pelajaran
predikat_per_mapel = df.groupby(["category", "season"]).size().reset_index(name="Jumlah")

# Membuat bar chart
plt.figure(figsize=(12, 6))
ax = sns.barplot(data=predikat_per_mapel, x="category", y="Jumlah", hue="season")

# Menambahkan nilai jumlah data di atas setiap batang
for p in ax.containers:
    ax.bar_label(p, fmt='%d', label_type='edge', fontsize=10, padding=3)

# Menambahkan judul dan label
plt.title("Distribusi Kategori berdasarkan Musim", fontsize=14)
plt.xlabel("Mata Pelajaran", fontsize=12)
plt.ylabel("Jumlah Item Kategori", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Musim")

# Menampilkan grafik
plt.show()
```
